## Clean

```python
# make a copy of the original DataFrame
df_clean = df.copy()
```

#### Fill Missing Values in homepage, tagline, overview, and release_date
##### Define
Replace missing values in these columns with appropriate placeholders.

##### Code

```python
df_clean['homepage'] = df_clean['homepage'].fillna("Unknown")
df_clean['tagline'] = df_clean['tagline'].fillna("No tagline")
df_clean['overview'] = df_clean['overview'].fillna("No overview available")
df_clean['release_date'] = df_clean['release_date'].fillna("1900-01-01")
```

##### Test

```python
df_clean[['homepage', 'tagline', 'overview', 'release_date']].isnull().sum()
```

#### Fill Missing Runtimes with Median Value
##### Define
Replace missing runtimes with the median runtime.

##### Code

```python
median_runtime = df_clean['runtime'].median()
df_clean['runtime'] = df_clean['runtime'].fillna(median_runtime)
```

##### Test

```python
df_clean['runtime'].isnull().sum()
```

#### Replace 0 in Budget and Revenue with Median of Non-Zero Values
##### Define
Since 0 likely means missing, replace it with the median of non-zero values.

##### Code

```python
median_budget = df_clean[df_clean['budget'] > 0]['budget'].median()
df_clean['budget'] = df_clean['budget'].replace(0, median_budget)

median_revenue = df_clean[df_clean['revenue'] > 0]['revenue'].median()
df_clean['revenue'] = df_clean['revenue'].replace(0, median_revenue)
```

##### Test

```python
df_clean[df_clean['budget'] == 0]['budget'].sum(), df_clean[df_clean['revenue'] == 0]['revenue'].sum()
```

#### Convert release_date to Datetime
##### Define
Change release_date from string to proper datetime format.

##### Code

```python
df_clean['release_date'] = pd.to_datetime(df_clean['release_date'])
```

##### Test

```python
df_clean['release_date'].dtypes
```

#### Clean JSON-Like String Columns into Lists of Names
##### Define
Convert stringified lists of dictionaries in specific columns into lists of names using ast.literal_eval().

##### Code

```python
json_like_cols = ['genres', 'production_companies', 'production_countries', 
                  'spoken_languages', 'keywords']
```

```python
for col in json_like_cols:
    df_clean[col] = df_clean[col].apply(lambda x: [i['name'] for i in ast.literal_eval(x)] if isinstance(x, str) and x.strip() else [])
```

##### Test

```python
df_clean[json_like_cols].head()
```

#### Rename Columns for Clarity
##### Define
Rename columns for clearer, consistent naming.

```python
df_clean.rename(columns={
    'vote_average': 'avg_rating',
    'vote_count': 'rating_count',
    'release_date': 'release_date',  
}, inplace=True)
```
